# 2 - wrapper Python StackAPI 

```python
site = StackAPI('stackoverflow')
```

```python
questions = site.fetch('questions', fromdate=1514764800, todate=1546300800,
                       min=51, tagged='python', sort='votes', pagesize=10)
```

```python
for question in questions['items']:
    print("Titre : ", question['title'])
    print("Balises : ", question['tags'])
    print("Date de création : ", question['creation_date'])
    print("Score de votes : ", question['score'])
    print("\n")
```

```python

# liste pour stocker questions
questions_data = []

for question in questions['items']:
    title = question['title']
    tags = question['tags']
    creation_date = pd.to_datetime(question['creation_date'], unit='s')
    score = question['score']
    
    questions_data.append([title, tags, creation_date, score])

query_50_API = pd.DataFrame(questions_data, columns=['Title', 'Tags', 'CreationDate', 'Score'])


```
